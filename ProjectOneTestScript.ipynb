```python
from CRUD_Python_Module import AnimalShelter

# Create object
shelter = AnimalShelter()

# ------------------
# CREATE TEST
# ------------------

test_data = {
    "animal_id": "A999999",
    "name": "Reggie",
    "animal_type": "Dog",
    "breed": "German Shepherd",
    "color": "Black and Tan",
    "age_upon_outcome": "2 years"
}

create_result = shelter.create(test_data)

print("CREATE RESULT:")
print(create_result)

# ------------------
# READ TEST
# ------------------

print("\nREAD RESULT:")

read_result = shelter.read({"name": "Reggie"})

for record in read_result:
    print(record)

# ------------------
# UPDATE TEST
# ------------------

update_result = shelter.update(
    {"name": "Reggie"},
    {"color": "Brown"}
)

print("\nUPDATE RESULT:")
print(update_result)

updated_record = shelter.read({"name": "Reggie"})

for record in updated_record:
    print(record)

# ------------------
# DELETE TEST
# ------------------

delete_result = shelter.delete(
    {"name": "Reggie"}
)

print("\nDELETE RESULT:")
print(delete_result)

remaining_records = shelter.read({"name": "Reggie"})

print("\nAFTER DELETE:")

for record in remaining_records:
    print(record)
```
